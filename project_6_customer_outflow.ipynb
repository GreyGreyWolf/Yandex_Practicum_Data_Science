## Исследование задачи

**Принадлежность к категориям вычисляется по формулам, состоящим из признаков. Они могут быть только числовыми. В данных присутствуют и категориальные признаки(object),значит, при обучении моделей будет ошибка. Преобразуем категориальные данные в числовые с помощью техники прямого кодирования (OHE).Она подходит для DecisionTreeClassifier, RandomForestClassifier, LogisticRegression. А что бы не угодить в ловушку фиктивных признаков(обучение будет сложнее), так как данных в избытке, применим параметр drop_first=True(удалить первый столбец).**

```python
# Примени технику OHE к каждому полученному набору данных
data_without_nan = pd.get_dummies(data_without_nan, drop_first=True)
data_with_zero = pd.get_dummies(data_with_zero, drop_first=True)
data_with_median = pd.get_dummies(data_with_median, drop_first=True)
```

**Разобьем каждый новый набор данных на тренировочную, валидационную и тестовую выборки (в пропорции 3:1:1):**

- **На тренировочной - будем обучать модели**

- **На валидационной - подберем лучшие гиперпараметры**

- **На тестовой выборке оценим качество моделей.**

**Новых датасетов у меня 3, поэтому, что бы оптимизировать процесс,  я напишу функцию, которая будет получать на вход датасет и название целевого признака, а возвращать словарь с 3-мя наборами выборок с целевыми признаками для каждой.**

```python
def get_selections(df, target_attribute):
    cnt = 0
    dict_with_selections = {}
    list_with_new_df = []
    flag = ''
    df_train, common_for_valid_and_test = train_test_split(df, test_size=0.4, random_state=12345)
    list_with_new_df.append(df_train)
    df_valid, df_test = train_test_split(common_for_valid_and_test, test_size=0.5, random_state=12345)
    list_with_new_df.append(df_valid)
    list_with_new_df.append(df_test)
    for new_df in list_with_new_df:
        cnt += 1
        features = new_df.drop(target_attribute, axis=1)
        target = new_df[target_attribute]
        if cnt == 1:
            flag = 'train'
        elif cnt == 2:
            flag = 'valid'
        elif cnt == 3:
            flag = 'test'
        dict_with_selections[flag] = [features, target]
    return dict_with_selections
```

```python
selections_without_nan = get_selections(data_without_nan, 'exited')
selections_with_zero = get_selections(data_with_zero, 'exited')
selections_with_median = get_selections(data_with_median, 'exited')
```

```python
def get_selection_shape(lst_with_selections):
    cnt = 0
    for s in lst_with_selections:
        cnt += 1
        if cnt == 1:
            print('\n','=' * 10, 'DF_WITHOUT_NAN', '=' * 10)
        elif cnt == 2:
            print('\n', '=' * 10, 'DF_WITH_ZERO', '=' * 10)
        elif cnt == 3:
            print('\n','=' * 10, 'DF_WITH_MEDIAN', '=' * 10)
        for name_of_selection, list_with_features_and_train in s.items():
            print(f'\nРазмер выборки {name_of_selection} -', list_with_features_and_train[0].shape)
```

### Пороги.

**Для достижения лучшего качества модели необходимо подобрать оптимальный порог, при котором будет максимальное значение истинно положительных ответов (TPR), и минимально значение ложно положительных ответов(FPR).**

**Пороги будем перебирать в параметрах модели. Напишем функцию для построения графика ROC-кривой (кривая ошибок) на графике которой, кроме кривых для моделей каждого набора выборок, будет случайная модель,  по горизонтали будут значения FPR, а по вертикали TPR. Каждая точка на графике соответствует выбору некоторого порога. Кривая ошибок позволит оценить полученное соотношение TPR и FPR. В идеале значение FPR должно быть меньше 0.5, а TPR больше 0.5., а площадь под графиком(AUC-ROC) должна стремится к еденице.**

```python
def draw_graph_roc_curve(dct_models_param):
    best_f1 = 0
    all_thresholds = 0
    best_models_thresholds = 0
    max_val_tpr = 0
    min_val_fpr = 0
    best_auc_roc = 0
    
    def find_nearest_num(ar_for_search, val):
        idx = np.abs(ar_for_search - val).argmin()
        return ar_for_search.flat[idx]

    
    plt.figure(figsize=[15, 12])
    plt.plot([0,1],[0,1], linestyle ='--')
    for key in dct_models_param.keys():
        fpr = dct_models_param[key]['fpr']
        tpr = dct_models_param[key]['tpr']
        f1 = dct_models_param[key]['f1']
        if (f1 > best_f1):
            best_f1 = f1
            best_models_thresholds = dct_models_param[key]['threshold']
            all_thresholds = dct_models_param[key]['all_thresholds']
            best_auc_roc = dct_models_param[key]['auc_roc']
        plt.plot(fpr, tpr, label=f'{key}')
    nearest_num = find_nearest_num(all_thresholds, best_models_thresholds)
    a = np.where(all_thresholds == nearest_num)
    index = int(a[0])
    max_val_tpr = list(tpr)[index]
    min_val_fpr = list(fpr)[index]
    plt.text((min_val_fpr + 0.02),(max_val_tpr + 0.02),
             f'THRESHOLD - {best_models_thresholds:.2f}, FPR - {min_val_fpr:.2f}, TPR - {max_val_tpr:.2f}',
             color='r',
             fontsize=16)
    plt.plot([min_val_fpr],[max_val_tpr],'o', color='r', markersize=15)
    plt.axvline(x=min_val_fpr, color='r', linestyle='--')
    plt.axhline(y=max_val_tpr, color='r', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('FPR', fontsize=15)
    plt.ylabel('TPR', fontsize=15)
    plt.figtext((min_val_fpr + 0.2), (max_val_tpr - 0.02), 'BEST_AUC_ROC - %.2f' % best_auc_roc, fontsize=17)
    plt.title('ROC-кривая', fontsize=17)
    plt.grid(True)
    plt.legend()
    plt.show();
```

```python
def get_best_params_for_decision_tree(selection_train, selection_valid_or_test, value_weight=None):
    best_params_for_model = None
    best_f1_for_decision_tree = 0
    best_treshold = 0
    for depth in range(4,11, 1):
        for min_split in range(2,11, 1):
            for min_leaf in range(1,11, 1):
                for crit in ['gini', 'entropy']:
                    for split in ['best', 'random']:
                        if value_weight is None:
                            model_decision_tree = DecisionTreeClassifier(random_state=12345, 
                                                                         criterion=crit,
                                                                         splitter=split,
                                                                         max_depth=depth,
                                                                         min_samples_split=min_split,
                                                                         min_samples_leaf=min_leaf)
                        elif value_weight == 'balanced':
                            model_decision_tree = DecisionTreeClassifier(random_state=12345, 
                                                                         criterion=crit,
                                                                         splitter=split,
                                                                         max_depth=depth,
                                                                         min_samples_split=min_split,
                                                                         min_samples_leaf=min_leaf,
                                                                         class_weight=value_weight)                            
                        
                        model_decision_tree.fit(selection_train[0],
                                                selection_train[1])
                        probalities_valid = model_decision_tree.predict_proba(selection_valid_or_test[0])
                        probabilities_one_valid = probalities_valid[:, 1]
                        #Переберем пороги
                        for threshold in np.arange(0.1, 0.9, 0.1):
                            predicted_valid = probabilities_one_valid > threshold
                            f1 = f1_score(selection_valid_or_test[1], predicted_valid)
                            if (f1 > best_f1_for_decision_tree):
                                best_f1_for_decision_tree = f1
                                best_threshold = threshold
                        if (best_threshold >= 0.1 ):
                            min_boundary = (best_threshold - 0.05)
                            max_boundary = (best_threshold + 0.05)
                            for updated_threshold in np.arange(min_boundary, max_boundary, 0.01):
                                predicted_valid = probabilities_one_valid > updated_threshold
                                f1 = f1_score(selection_valid_or_test[1], predicted_valid)
                                if (f1 > best_f1_for_decision_tree):
                                    best_f1_for_decision_tree = f1
                                    best_threshold = updated_threshold
    # Добавим метрику AUC_ROC, чем выше будет показатель, тем лучше модель будет разделять случаи
    auc_roc = roc_auc_score(selection_valid_or_test[1], probabilities_one_valid)
    fpr, tpr, all_thresholds = roc_curve(selection_valid_or_test[1], probabilities_one_valid)                  
    best_params_for_model = {'criterion': crit, 
                             'splitter': split,
                             'max_depth': depth,
                             'min_samples_split': min_split,
                             'min_samples_leaf': min_leaf,
                             'class_weight': value_weight,
                             'f1': best_f1_for_decision_tree,
                             'auc_roc': auc_roc,
                             'fpr': fpr,
                             'tpr': tpr,
                             'all_thresholds':all_thresholds,
                             'threshold': best_threshold}
    return best_params_for_model        
```

```python
def get_best_params_for_logistic_regression(selection_train, selection_valid_or_test, value_weight=None):
    best_params_for_model = None
    best_f1_for_logistic_regression = 0
    best_threshold = 0
    if value_weight is None:
        model_logistic_regression = LogisticRegression(random_state=12345, solver='liblinear')
    elif value_weight == 'balanced':
        model_logistic_regression = LogisticRegression(random_state=12345, solver='liblinear', class_weight=value_weight)
    model_logistic_regression.fit(selection_train[0], selection_train[1])    
    probalities_valid = model_logistic_regression.predict_proba(selection_valid_or_test[0])
    probabilities_one_valid = probalities_valid[:, 1]
    #Переберем пороги
    for threshold in np.arange(0.1, 0.9, 0.1):
        predicted_valid = probabilities_one_valid > threshold
        f1 = f1_score(selection_valid_or_test[1], predicted_valid, average='binary')
        if (f1 > best_f1_for_logistic_regression):
            best_f1_for_logistic_regression = f1
            best_threshold = threshold
    if (best_threshold >= 0.1 ):
        min_boundary = best_threshold - 0.05
        max_boundary = best_threshold + 0.05
        for updated_threshold in np.arange(min_boundary, max_boundary, 0.01):
            predicted_valid = probabilities_one_valid > updated_threshold
            f1 = f1_score(selection_valid_or_test[1], predicted_valid)
            if (f1 > best_f1_for_logistic_regression):
                best_threshold = updated_threshold
                best_f1_for_logistic_regression = f1
    # Добавим метрику AUC_ROC, чем выше будет показатель,
    # тем лучше модель будет разделять случаи
    auc_roc = roc_auc_score(selection_valid_or_test[1], probabilities_one_valid)
    fpr, tpr, all_thresholds = roc_curve(selection_valid_or_test[1], probabilities_one_valid)
    best_params_for_model = {'class_weight': value_weight,
                             'f1': best_f1_for_logistic_regression, 
                             'auc_roc': auc_roc,
                             'fpr': fpr,
                             'tpr': tpr,
                             'all_thresholds': all_thresholds,
                             'threshold': best_threshold}                                                 
    return best_params_for_model
```

**Вместо повторения важных объектов, уберем часть признаков, которых больше.**

```python
def downsample(features, target):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=0.1, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=0.1, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return {'train':[features_downsampled, target_downsampled]}
```

```python
def get_result_logisitic_regression(selection_train, selection_valid_or_test, param):
    model = LogisticRegression(random_state=12345, 
                               solver='liblinear',
                               class_weight=param['class_weight'])
    model.fit(selection_train[0], selection_train[1])
    probalities_valid = model.predict_proba(selection_valid_or_test[0])
    probabilities_one_valid = probalities_valid[:, 1]        
    threshold = param['threshold']
    predictions = model.predict(selection_valid_or_test[0])
    f1 = f1_score(selection_valid_or_test[1], predictions, average='binary')
    auc_roc = roc_auc_score(selection_valid_or_test[1], probabilities_one_valid)
    fpr, tpr, all_thresholds = roc_curve(selection_valid_or_test[1], probabilities_one_valid)
    result = {'f1': f1, 
              'auc_roc': auc_roc,
              'threshold': threshold,
              'all_thresholds': all_thresholds,
              'fpr': fpr,
              'tpr': tpr}
    return result
```

```python
def get_result_random_forest(selection_train, selection_valid_or_test, param):
    model = RandomForestClassifier(random_state=12345, 
                                   class_weight=param['class_weight'], 
                                   n_estimators=param['n_estimators'],
                                   max_depth=param['max_depth'])  
    model.fit(selection_train[0], selection_train[1])
    probalities_valid = model.predict_proba(selection_valid_or_test[0])
    probabilities_one_valid = probalities_valid[:, 1]        
    threshold = param['threshold']
    predictions = model.predict(selection_valid_or_test[0])
    f1 = f1_score(selection_valid_or_test[1], predictions, average='binary')
    auc_roc = roc_auc_score(selection_valid_or_test[1], probabilities_one_valid)
    fpr, tpr, all_thresholds = roc_curve(selection_valid_or_test[1], probabilities_one_valid)
    result = {'f1': f1, 
              'auc_roc': auc_roc,
              'threshold': threshold,
              'all_thresholds': all_thresholds,
              'fpr': fpr,
              'tpr': tpr}
    return result
```

```python
def get_result_decision_tree(selection_train, selection_valid_or_test, param):
    model = DecisionTreeClassifier(random_state=12345, 
                                   criterion=param['criterion'],
                                   splitter=param['splitter'],
                                   max_depth=param['max_depth'],
                                   min_samples_split=param['min_samples_leaf'],
                                   min_samples_leaf=param['min_samples_leaf'],
                                   class_weight=param['class_weight'])   
    model.fit(selection_train[0], selection_train[1])
    probalities_valid = model.predict_proba(selection_valid_or_test[0])
    probabilities_one_valid = probalities_valid[:, 1]        
    threshold = param['threshold']
    predictions = model.predict(selection_valid_or_test[0])
    f1 = f1_score(selection_valid_or_test[1], predictions, average='binary')
    auc_roc = roc_auc_score(selection_valid_or_test[1], probabilities_one_valid)
    fpr, tpr, all_thresholds = roc_curve(selection_valid_or_test[1], probabilities_one_valid)
    result = {'f1': f1, 
              'auc_roc': auc_roc,
              'threshold': threshold,
              'all_thresholds': all_thresholds,
              'fpr': fpr,
              'tpr': tpr}
    return result
```
