## Предобработка данных

Исправим названия столбцов ***cityCenters_nearest, parks_around3000, ponds_around3000*** на корректные ( соответствующие формату snake_case)

```python
data = data.rename(columns = {'cityCenters_nearest' : 'city_centers_nearest', 'parks_around3000': 'parks_around_3000', 'ponds_around3000': 'ponds_around_3000'})
```

```python
data.columns
```

**Теперь посмотрим на пропуски и их количетво.**

```python
data.isna().sum()
```

**Последовательно исключим из массива данных строки, в которых есть пропуски для столбцов "Населенный пункт" и   "Всего этажей в доме", так как их количество не велико , и не повлияет на дальнейшие расчеты.**

```python
data.dropna(subset = ['locality_name'], inplace=True)
data.dropna(subset = ['floors_total'], inplace=True)
```

```python
print('Количесво пропусков в столбце "Название населенного пункта" - ', data['locality_name'].isna().sum())
print('\nКоличесво пропусков в столбце "Количество этажей в доме" - ', data['floors_total'].isna().sum())
```

**В случае с количеством балконов, пропуски могут означать, что балконов нет. Не смотря на информацию о количестве комнат и размере жилой площади, количестве этажей, а так же расстоянии до центра, что может предпологать, что дома находятся в одном районе - вовсе не означает, что все дома типовой застройки. Тогда имеет смысл заменить пропуски на 0.**

```python
data['balcony'] = data['balcony'].fillna(0)
```

**Для пропущенных значений в стоблце "жилая площадь", при условии, что квартира является студией, жилой площадью (полезной площадью) будет считаться вся площадь квартиры.**

```python
def get_living_area_for_studio(row):
    if (row['studio'] == True) and (pd.isna(row['living_area'])):
        return row['total_area']

print('Количество отсутствующих значений в столбце "Жилая площадь" до обработки', data['living_area'].isna().sum())
data.loc[(data['studio'] == True) & (data['living_area'].isna()), 'living_area'] = data.apply(get_living_area_for_studio, axis = 1)
print('\nКоличество отсутствующих значений в столбце "Жилая площадь" до обработки', data['living_area'].isna().sum())
```

***Жилая площадь - это сумма площадей жилых комнат, предназначенных проектом для постоянного проживания (спален, гостиных).***

**Для квартир не являющихся студиями вычислим приблизительную жилую площадь (медиану) на основе информации  о количестве комнат и общей площади. В случае, если не будет группировки по этим значения, возьмем медиану по значениям жилой площади в каждом населенном пункте.**

```python
def get_living_area_for_not_studio(row):
    if (row['studio'] == False) and (pd.isna(row['living_area'])):
        m = data[(data['total_area'] == row['total_area']) & (data['rooms'] == row['rooms'])]['living_area'].median()
        if pd.isna(m):
            m = data[(data['locality_name'] == row['locality_name'])]['living_area'].median()  
        return m


data.loc[(data['studio'] == False) & (data['living_area'].isna()), 'living_area'] = data.apply(get_living_area_for_not_studio, axis=1)
```

```python
print('Количество отсутствующих значений в столбце "Жилая площадь" - ', data['living_area'].isna().sum())
```

**Для пропущенных значений в стоблце "Высота потолков" возьмем медианное значение высоты потолков квартир осортированных по кличеству этажей в жилом доме. В случае, если не будет группировки по этому значению, возьмем медиану по значениям высоты потолков в каждом населенном пункте.**

```python
def get_ceiling_height(row):
    if (pd.isna(row['ceiling_height'])):
        m = data[data['floors_total'] == row['floors_total']]['ceiling_height'].median()
        if pd.isna(m):
            m = data[data['locality_name'] == row['locality_name']]['ceiling_height'].median()
        return m


data.loc[data['ceiling_height'].isna(), 'ceiling_height'] = data.apply(get_ceiling_height, axis=1)
```

```python
print('Количество отсутствующих значений в столбце "Высота потолков" - ', data['ceiling_height'].isna().sum())
```

**Для заполнения пропусков в столбце "Площадь кухни" используем:**

 - **Если квартира - не студия, то значение, полученное вычетанием жилой площади квартиры из общей площади**
 
 
 - **Если квартира - студия, то площадь кухни будет 0, так как в таком случае все пространство будет общим.**

```python
def get_kitchen_area(row):
    if (row['studio'] == False) and (pd.isna(row['kitchen_area'])):
        k = row['total_area'] - row['living_area']
        return k
    elif (row['studio'] == True) and (pd.isna(row['kitchen_area'])):
        return 0


data.loc[data['kitchen_area'].isna(), 'kitchen_area'] = data.apply(get_kitchen_area, axis=1)
```

```python
print('Количество отсутствующих значений в столбце "Площадь кухни" - ', data['kitchen_area'].isna().sum())
```

```python
df_with_e = data['locality_name'].str.contains('посёлок')
data.loc[df_with_e, 'locality_name'] = data.loc[df_with_e, 'locality_name'].str.replace('посёлок','поселок')
print('Количество слов "посёлок" написанного через "ё" - ',data['locality_name'].str.contains('посёлок').sum())
```
